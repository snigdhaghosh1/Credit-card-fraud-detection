```python
#Importing the dependencies
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the file
import os
os.getcwd()
creditcard = pd.read_csv("creditcard.csv")
```

```python
creditcard
```

```python
#first 5 rows of the dataset
creditcard.head()
```

```python
#check the number of missing values in each column.
creditcard.isnull().sum()
```

```python
# check the distribution of legit and fraudulent transactions
creditcard ['Class'].value_counts()
```

```python
# Separating the data for analysis
legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
#the mean value here is more than the legit value mean. 
```

```python
#compare the values for both transactions
creditcard.groupby('Class').mean()
```

```python
#Take random samples from the data to equalise
legit_sample = legit.sample(n=492)
```

```python
#concatenate the two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
#check first 5 rows of new data
new_dataset.head()
```

```python
#check last 5 rows of new data
new_dataset.tail()
```

```python
#to check the value counts
new_dataset['Class'].value_counts()
```

```python
#Get the new means of the new data
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into Features and Targets.
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
#Print the data
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Train the model using LogisticRegression
model = LogisticRegression ()
```

```python
#Training the logisticregression model with training data
model.fit(X_train, Y_train)
```

```python
#Performance or evaluation of the model based on accuracy score:
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data:', training_data_accuracy)
```
